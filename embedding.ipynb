```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences
sent = [
    'the glass of milk',
    'the cup of tea',
    'i m a good boy',
    'i m a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
# define the vocab size
vocab_size = 10000
```

```python
# one hot representation
one_hot_rep = [one_hot(words, vocab_size)for words in sent ]
```

```python
one_hot_rep
```

```python
# word enbedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
# from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_doc = pad_sequences(one_hot_rep, padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
#  feature representation
dim=10
```

```python
model = Sequential()
model.add(Embedding(vocab_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
